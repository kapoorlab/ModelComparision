```python
import sys
import os
os.environ["CUDA_VISIBLE_DEVICES"]="1"
import glob
import numpy as np
import csv
from tifffile import imread
sys.path.append("../AnalyzeCSV")
from CreateImages import TimelapseImage, TwoDImage, CountTwoDevents, CountTimeLapseevents
try:
    from pathlib import Path
    Path().expanduser()
except (ImportError,AttributeError):
    from pathlib2 import Path

try:
    import tempfile
    tempfile.TemporaryDirectory
except (ImportError,AttributeError):
    from backports import tempfile
```

```python
#Event File Append names
Mature = 'CNNMatureLocation'
NonMature = 'CNNNonMatureLocation'
Macrocheate = 'CNNMacroKittyLocation'

MatureEventCounter = []
NonMatureEventCounter = []
MacroEventCounter = []
```

```python
save_dir = '/home/sancere/Desktop/pzKO3a/Results/'
csv_dir = '/home/sancere/NextonDisk_1/StaticpzKO3a/'
image_dir =  '/home/sancere/Desktop/pzKO3a/'


Raw_path = os.path.join(image_dir,'*.tif')
X = glob.glob(Raw_path)




Path(save_dir).mkdir(exist_ok = True)


for image_file in X:

  image= imread(image_file)
  image = np.asarray(image)
  Name = os.path.basename((os.path.splitext(image_file)[0])) 

  #Macocheate Events 
  csv_Name = Macrocheate + Name
  csv_file = csv_dir + csv_Name + '.csv'
 
  if os.path.exists(csv_file):

      emptyimage = np.zeros([image.shape[0], image.shape[1]], dtype='uint16')
      TwoDImage(csv_file, emptyimage, csv_Name, save_dir)
      MacroCount, Label = CountTwoDevents(csv_file, csv_Name)  
      MacroEventCounter.append([MacroCount, Label])
      
  #Mature Events 
  csv_Name = Mature + Name
  csv_file = csv_dir + csv_Name + '.csv'

  if os.path.exists(csv_file):

      emptyimage = np.zeros([image.shape[0], image.shape[1]], dtype='uint16')
      TwoDImage(csv_file, emptyimage, csv_Name, save_dir)
      MatureCount, Label = CountTwoDevents(csv_file, csv_Name)  
      MatureEventCounter.append([MatureCount, Label])  
   
  #NonMature Events
  csv_Name = NonMature + Name
  csv_file = csv_dir + csv_Name + '.csv'

  if os.path.exists(csv_file):

      emptyimage = np.zeros([image.shape[0], image.shape[1]], dtype='uint16')
      TwoDImage(csv_file, emptyimage, csv_Name, save_dir)
      NonMatureCount, Label = CountTwoDevents(csv_file, csv_Name)  
      NonMatureEventCounter.append([NonMatureCount, Label]) 
    
        
for line in MacroEventCounter:
    writer = csv.writer(open(save_dir + "Macrocheate" +".csv", "w"))
    writer.writerows(MacroEventCounter)
   

for line in MatureEventCounter:
    writer = csv.writer(open(save_dir + "Mature" +".csv", "w"))
    writer.writerows(MatureEventCounter)
    
for line in NonMatureEventCounter:
    writer = csv.writer(open(save_dir + "NonMature" +".csv", "w"))
    writer.writerows(NonMatureEventCounter)    
```
